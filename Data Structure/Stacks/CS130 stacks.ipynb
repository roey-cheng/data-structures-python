# stack -- a linear data structure
## Last in, First out Property (ie, add at top, remove from top)
### in python, use a list to represent a stack

### Whenever you learn a new Abstract Data Type (ADT), ask:
Can an old ADT help implement it efficiently?
Can this new ADT help improve existing implementations?
Does the problem I’m solving resemble another one I’ve solved before?


### operations can be used with a Stack Abstract Data:
- Create an empty stack                Stack()
- Determine weather a stack is empty   is_empty()
- Push (add a new item to the stack)    push(item)
- pop (remove from the stack the item that was added most recently)     pop()
- peek (Retrieve from the stack the item that was added most recently)  peek()
- size() returns the number of items on the stack

```python
class Stack:
    def __init__(self):
        self.items = []
    def is_empty(self):
        return self.items == []
    def size(self):
        return len(self.items)
    def push(self, item):
        self.items.append(item) # O(1)
    def pop(self):
        return self.items.pop() # O(1)
    



    def another_push(self, item):
        self.items.insert(0, item) #O(n)
    def another_pop(self):
        self.pop(0) #O(n)
    def double_push(self, element):
        self.push(element)
        self.push(element)
    def join(self):
        result = ''
        for i in range(self.items.size()):
            result += str(self.items.pop())
        return result
```
